```python
Year = input("Enter the year ")
sector = input("Enter the Sector ")
sector=sector.upper()
sec = np.array(data['Sector'].where(Sector==sector))
year = np.array(data[Year].where(Sector==sector))
info=[]

for i in sec:
    if i==sector:
        info.append(i)
for i in year:
    if i>=0:
        info.append(i)
for i in range(0,2):
    print(info[i],end=" ")
print(type(info[1]))
```

```python
year = input("Enter Year ")
print("Mean ",data[year].mean())
print("Median ",data[year].median())
print("Mode ",data[year].mode())
```

```python
Years=[]
for i in data.columns:
    if i!='Sector':
        Years.append(i)
Years=np.array(Years)
Years
```

```python
sector = input("Enter Sector ")
sector=sector.upper()
Total=[]

for i in Years:
    Total.append(data[i].where(Sector==sector))

Total=np.array(Total)
sum = []
for i in Total:
    for j in i:
        if j>=0:
            sum.append(j)
sum=np.array(sum)
print("Total investment ",sum.sum())
print("Mean of Investement ",sum.mean())
print("Minimun Investment ",sum.min())
print("Maximun Investment ",sum.max())
```
